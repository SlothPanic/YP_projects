### Сколько игр выпускалось в разные годы

```python
##########Функция для создания сводной таблицы##########
def pivot(
    name,
    df, 
    index, 
    values, 
    aggfunc):
    print('..........Сводная таблица {}..........'.format(name))
    pivot = df.pivot_table(index=index, 
                   values=values,
                   aggfunc=aggfunc
                  )
    return pivot
#########################################################
##########Функция для создания статистических данных##########
def stat(
    name, 
    df,
    variable_1, 
    variable_2):
    print('..........Статистические данные {}..........'.format(name))
    stat=df.describe(percentiles=[.01,.25,.50,.99]).T
    stat['sum'] = df.sum()
    stat['var'] = [round(np.var(variable_1),2),
                        round(np.var(variable_2),2)]
    stat['std_dev'] = [round(np.std(variable_1),2),
                            round(np.std(variable_2),2)]
    return stat
    print('')
    print('')
##############################################################
##########Функция для создания гистограмм##########
def hist(
    name,
    label_y, 
    label_x, 
    variable_1, 
    variable_2):
    print('..........{}..........'.format(name))
    plt.figure(figsize=(17,8))
    plt.title('Зависимость: {} - {}'.format(label_y, label_x))
    plt.xlabel('Год выпуска')
    plt.ylabel('Количество игр')
    plt.grid(True)
    plt.plot(variable_1, variable_2,'o-')
    plt.axhline(variable_2.mean(), 
                color='r',linestyle='dashed',linewidth=2,label='Mean number: {:.2f}'.format(variable_2.mean()))
    plt.axhline(variable_2.median(), 
                color='g',linestyle='dashed',linewidth=2,label='Median number: {:.0f}'.format(variable_2.median()))
    plt.axhline(variable_2.max(), 
                color='y',linestyle='solid',linewidth=2,label='Max number: {:.0f}'.format(variable_2.max()))
    plt.axhline(variable_2.min(), 
                color='k',linestyle='solid',linewidth=2,label='Min number: {:.0f}'.format(variable_2.min()))
    plt.legend()
    plt.show()
    ###################################################
```

```python
game_per_year = pivot(
    'выпущенных игр в год',
    df, 
    'year_of_release', 
    'name', 
    'count').reset_index()
game_per_year.columns = ['year_of_release','number']
display(game_per_year.sort_values('number'))
print('')
print('')
print('Минимальное количество игр выпущено в 1980 году:', 
      game_per_year['number'].min())
print('')
print('Максимальное количество игр выпущено в 2008 году:', 
      game_per_year['number'].max())
print('')
print('')

game_per_year_stat = stat('выпущенных игр в год', 
                          game_per_year, 
                          game_per_year['year_of_release'], 
                          game_per_year['number'])
display(game_per_year_stat)
print('')
print('')

hist('График зависимости выпущенных игр в год',
     'количества выпущенных игр', 
     'год', 
     game_per_year['year_of_release'], 
     game_per_year['number'])
```

## Проверка гипотез

На данном этапе проверим следующие гипотезы:
1. Средние пользовательские рейтинги платформ **Xbox One(XOne)** и **PC** одинаковые.
2. Средние пользовательские рейтинги жанров **Action** и **Sports** разные.

### Гипотеза №1: Средние пользовательские рейтинги платформ Xbox One(XOne) и PC одинаковые

Обозначим двустороннюю гипотезу по заданным условиям:
- **Нулевая гипотеза (H0)** - Средние пользовательские рейтинги платформ Xbox One(XOne) и PC - **равны**.
- **Альтернативная гипотеза (H1)** - Средние пользовательские рейтинги платформ Xbox One(XOne) и PC - **не равна**.

Подготовим выборки для проверки гипотезы:

```python
df_after_2013.dropna(subset=['user_score'], inplace=True)
print('..........Выборка для Xbox One..........')
xone_df = df_after_2013.query('platform == "XOne"')
display(xone_df)
print('Дисперсия составила:', round(
    np.var(xone_df['user_score'], ddof=1),2))
print('Стандартное отклонение:', round(
    np.std(xone_df['user_score'], ddof=1),2))
print('')
print('')
print('..........Выборка для PC..........')
pc_df = df_after_2013.query('platform == "PC"')
display(pc_df)
print('Дисперсия составила:', round(
    np.var(pc_df['user_score'], ddof=1),2))
print('Стандартное отклонение:', round(
    np.std(pc_df['user_score'], ddof=1),2))
```

По условиям проверки у нас даны 2 случайные выборки из пользовательских оценок для двух разных платформ. В выборках даны более 30 элементов, значит выборочные среднее и дисперсия близки к реальному среднему и дисперсии генеральной совокупности.

Дисперсии выборок сильно отличается, стандартное отклонение близко по значению.

Проводим проверку нашей гипотезы при помощи **"t-теста"**, критический уровень статистической значимости примем **0,05**:

```python
def hypothesis(alpha, sample_1, sample_2):
    alpha = alpha
    hyp = st.ttest_ind(
    sample_1, 
    sample_2
    )

    print('p-значение: ', hyp.pvalue)
    
    if hyp.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')

hypothesis(.05, xone_df['user_score'], pc_df['user_score'])
```

#### Вывод по гипотезе №1: Средние пользовательские рейтинги платформ Xbox One(XOne) и PC одинаковые

Значение р-value достаточно велико для того, чтобы отвергать нулевую гипотезу: ~ 25,2%. Для доказательства мы взяли 5% критический уровень статистической значимости, куда значение не попало. Делаем вывод:
- **Нулевая гипотеза (H0)** не отвергнута, принимаем ее как вероятное доказательство гипотезы: Средние пользовательские рейтинги платформы Xbox One(XOne) **равны** средним пользовательским рейтингам платформы PC. Разница между оценками пользователей Xbox One и PC не настолько велика.
