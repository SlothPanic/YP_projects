Согласно прилагаемой документации, в нашем распоряжении 3 файла формата `.csv` со следующим содержимым:

   - **`revenue`** — выручка заказа
   - **`group`** — группа A/B-теста, в которую попал заказ
   
3. Файл `visitors.csv` — количество пользователей в каждую дату в указанной группе для A/B теста

**Структура visitors.csv:**

## Приоритизация гипотез:

1. Применение фреймворка ICE для приоритизации гипотез. Сортировака по убыванию приоритета.
2. Применение фреймворка RICE для приоритизации гипотез. Сортировака по убыванию приоритета.
3. Указать, изменение приоритизации гипотиз при использовании разных фреймворков.

### Применение фреймворка ICE для приоритизации гипотез

Один из самых популярных методов приоритизации является **`ICE`**. Его составляющие `impact`, `confidence`, `effort` / ease («влияние, уверенность, усилия / простота»). Применим его для получения приоритизации имеющихся у нас гипотез:

```python
hypothesis['ICE'] = hypothesis['impact']*hypothesis['confidence']/hypothesis['efforts']
hypothesis.sort_values('ICE', ascending=False)
```

### Применение фреймворка RICE для приоритизации гипотез

У метода **`ICE`** есть модификация **`RICE`**. К его составляющим добавляется значение "охвата"(скольких пользователей охватит изменение) - `reach`. Применим его для получения приоритизации гипотез:

```python
hypothesis['RICE'] = hypothesis['reach']*hypothesis['impact']*hypothesis['confidence']/hypothesis['efforts']
hypothesis.sort_values('RICE', ascending=False)
```

### Изменение приоритизации гипотиз при использовании разных фреймворков.

По изученным данным, получены данные и разночтения по приоритизации гипотез:
- приоритет гипотез после применения метода **`ICE`**: первые три приоритетные задачи **№8, №0, №7**
- с учетом дополнительного фактора **охвата пользователей** применяя метод **`RICE`** приоритет гипотез поменялся: первые три задачи **№7, №2, №0**
- гипотеза **№8** опустилась в середину списка и стала не приоритетной
- применяя метод **`RICE`** мы учитываем большие характеристики гипотез, следовательно, **результаты приоритизации более корректные** в нашем случае. При отсутствии параметра **охвата** первостепенным остается метод **`ICE`**

```python
y = hypothesis['ICE'] 
z = hypothesis['RICE'] 
n = hypothesis.index 

fig, ax = plt.subplots()
ax.scatter(z, y)

for i, txt in enumerate(n):
    ax.annotate(txt, (z[i], y[i]))

plt.title("ICE and RICE гипотезы") 
plt.xlabel("RICE") 
plt.ylabel("ICE");
```

### Построение графика кумулятивной выручки по группам

Создадим датафрейм с уникальными значениями дат и групп исследования таблицы исследования:

```python
dates_groups = orders[['date','group']].drop_duplicates()
```

Соберем агрегированные кумулятивные по дням данные о заказах. В новый датафрейм занесем:
- дату, 
- группу теста, 
- число уникальных заказов по указанную дату включительно, 
- число уникальных пользователей, совершивших хотя бы 1 заказ в группе теста по указанную дату включительно,
- суммарную выручку заказов в группе теста по указанную дату включительно

```python
orders_aggregated = dates_groups.apply(
    lambda x:orders[np.logical_and(
        orders['date'] <= x['date'], 
        orders['group'] == x['group']
    )
                   ]
.agg({'date':'max', 
      'group':'max', 
      'transaction_id':'nunique', 
      'visitor_id':'nunique', 
      'revenue':'sum'}
    ), axis=1).sort_values(by=['date','group'])
orders_aggregated.sample(5)
```

Аналогично получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина. В новый датафрейм занесем:
- дату, 
- группу теста, 
- количество уникальных посетителей в группе теста по указанную дату включительно.

```python
visitors_aggregated = dates_groups.apply(
    lambda x:visitors[np.logical_and(
        visitors['date'] <= x['date'], 
        visitors['group'] == x['group']
    )
                   ]
.agg({'date':'max', 
      'group':'max', 
      'visitors':'sum'}
    ), axis=1).sort_values(by=['date','group'])
visitors_aggregated.sample(5)
```

Объединим агрегированные данные в переменную **`camulative_data`** и переименуем столбцы соответственно их содержанию:

```python
camulative_data = orders_aggregated.merge(
    visitors_aggregated, 
    left_on=['date','group'], 
    right_on=['date','group']
)
camulative_data.columns = ['date', 
                           'group', 
                           'orders', 
                           'buyers', 
                           'revenue', 
                           'visitors']
camulative_data.head()
```

Для построения графиков кумулятивной выручки по группам по дням разделим данные на две группы и визуализируем их:

```python
cumulative_revenue_a = camulative_data[
    camulative_data['group'] == 'A'][['date','revenue','orders']]
cumulative_revenue_b = camulative_data[
    camulative_data['group'] == 'B'][['date','revenue','orders']]

plt.figure(figsize=(17,8))
plt.title('Графики кумулятивной выручки по дням и группам тестирования')
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.grid(True)
plt.plot(
    cumulative_revenue_a['date'], 
    cumulative_revenue_a['revenue'], 
    'o-',
    color='r',
    label='A')
plt.plot(
    cumulative_revenue_b['date'], 
    cumulative_revenue_b['revenue'], 
    'o-',
    color='k',
    label='B')
plt.legend()
plt.show()
```

#### Обоснование выводов и предложений:

Отметим следующие закономерности:
- до середины теста **сегмент B превышал сегмент A** не более чем на **50000** 
- 18.08.2019 наблюдается резкий рост и **превышение сегмента B** примерно на **180000**
- вероятно, **аномально большие заказы влияют на результаты**
- при заключительном анализе необходимо **исключить аномальные значения**

### Определить статистическую значимость различий в конверсии между группами по "сырым" данным

Первоначально определим статистическую значимость различий в конверсии и среднем чеке между группами по данным без удаления аномалий.

Обозначим двустороннюю гипотезу по заданным условиям:
- **Нулевая гипотеза (H0)** - Конверсии групп A и B **равны**.
- **Альтернативная гипотеза (H1)** - Конверсии групп A и B - **не равны**.

Выведем необходимые данные для проведения сравнительного анализа:

```python
visitors_a_daily = visitors[visitors['group'] == 'A'][['date','visitors']]
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

visitors_a_cummulative = visitors_a_daily.apply(
    lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']].agg(
        {'date':'max',
         'visitors_per_date_a':'sum'}
    ), axis=1
)
visitors_a_cummulative.columns = ['date', 'visitors_cummulative_a']

visitors_b_daily = visitors[visitors['group'] == 'B'][['date','visitors']]
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

visitors_b_cummulative = visitors_b_daily.apply(
    lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']].agg(
        {'date':'max',
         'visitors_per_date_b':'sum'}
    ), axis=1
)
visitors_b_cummulative.columns = ['date', 'visitors_cummulative_b']
```

```python
orders_a_daily = orders[orders['group'] == 'A'][['date',
                                                 'transaction_id',
                                                 'visitor_id',
                                                 'revenue']].groupby(
    'date', as_index=False).agg(
    {'transaction_id': pd.Series.nunique, 
     'revenue': 'sum'})
orders_a_daily.columns = ['date', 'order_per_date_a', 'revenue_per_date_a']

orders_a_cummulative = orders_a_daily.apply(
    lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']].agg(
        {'date':'max',
         'order_per_date_a':'sum',
         'revenue_per_date_a':'sum'}), axis=1).sort_values('date')
orders_a_cummulative.columns = ['date',
                                'orders_cummulative_a',
                                'revenue_cummulative_a',]

orders_b_daily = orders[orders['group'] == 'B'][['date',
                                                 'transaction_id',
                                                 'visitor_id',
                                                 'revenue']].groupby(
    'date', as_index=False).agg(
    {'transaction_id': pd.Series.nunique, 
     'revenue': 'sum'})
orders_b_daily.columns = ['date', 'order_per_date_b', 'revenue_per_date_b']

orders_b_cummulative = orders_b_daily.apply(
    lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']].agg(
        {'date':'max',
         'order_per_date_b':'sum',
         'revenue_per_date_b':'sum'}), axis=1).sort_values('date')
orders_b_cummulative.columns = ['date',
                                'orders_cummulative_b',
                                'revenue_cummulative_b',]
```

```python
data = orders_a_daily.merge(
    orders_b_daily, left_on='date', right_on='date', how='left').merge(
    orders_a_cummulative, left_on='date', right_on='date', how='left').merge(
    orders_b_cummulative, left_on='date', right_on='date', how='left').merge(
    visitors_a_daily, left_on='date', right_on='date', how='left').merge(
    visitors_b_daily, left_on='date', right_on='date', how='left').merge(
    visitors_a_cummulative, left_on='date', right_on='date', how='left').merge(
    visitors_b_cummulative, left_on='date', right_on='date', how='left')
data.head()
```

```python
orders_by_users_a = orders[orders['group'] == 'A'].groupby(
    'visitor_id', as_index=False).agg(
    {'transaction_id':pd.Series.nunique}
)
orders_by_users_a.columns = ['visitor_id','orders']

orders_by_users_b = orders[orders['group'] == 'B'].groupby(
    'visitor_id', as_index=False).agg(
    {'transaction_id':pd.Series.nunique}
)
orders_by_users_b.columns = ['visitor_id','orders']
```

```python
sample_a = pd.concat(
    [orders_by_users_a['orders'], 
     pd.Series(
         0, 
         index=np.arange(
             data['visitors_per_date_a'].sum() - len(
                 orders_by_users_a['orders']
             )
         ), 
         name = 'orders'
     )], 
    axis=0
)
sample_b = pd.concat(
    [orders_by_users_b['orders'], 
     pd.Series(
         0, 
         index=np.arange(
             data['visitors_per_date_b'].sum() - len(
                 orders_by_users_b['orders']
             )
         ), 
         name = 'orders'
     )], 
    axis=0
)
```

Применим тест Манна-Уитни:

```python
def hypothesis(alpha, sample_1, sample_2):
    alpha = alpha
    p_value = st.mannwhitneyu(
    sample_1, 
    sample_2
    )[1]

    print('p-value: ', round(p_value, 5))
    
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
        
    print('Относительное различие между группами: {0:.3f}'.format(
        sample_2.mean()/sample_1.mean() - 1))

hypothesis(.05, sample_a, sample_b)
```

#### Обоснование выводов и предложений:

Получены следующие результаты:
- **Нулевая гипотеза (H0)** - **отвергнута** (P-value < 0.05)
- анализ **"сырых данных"** показывает, что в конверсии между группами **есть статистически значимые различия**. Относительный **прирост группы B** к конверсии **группы А** равен **13,8%** - принимаем **Альтернативную гипотезу (H1)**

### Определить статистическую значимость различий в среднем чеке заказа между группами по "сырым" данным

По напиманным функциям в предыдущем шаге, проверим статистическую значимость различий в среднем чеке заказа между группами:
